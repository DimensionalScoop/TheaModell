```python

a = 20070301
b = 20071201

def time_mask(arr):
    #print(arr)
    #print(arr[(arr>=a & arr<b)])
    return arr[np.logical_and(arr>=a, arr<b)]#arr[(arr<a)|(b<=arr)]
```

```python
from joblib import Memory
memory = Memory("/tmp/joblib", verbose=0)

@memory.cache
def count_peaks(arr, peak_size=50):
    peaks = maximum_filter(arr, size=peak_size)
    return np.sum(arr==peaks)

#@memory.cache
def local_maxima(arr, peak_size=50):
    peaks = maximum_filter(arr, size=peak_size)
    return arr[np.where(arr==peaks)]

#@memory.cache
def local_minima(arr, peak_size=50):
    peaks = minimum_filter(arr, size=peak_size)
    return arr[np.where(arr==peaks)]
```

```python
@memory.cache
def grad_strength(pressure_map):
    x,y = np.gradient(pressure_map)
    return np.sqrt(x**2 + y**2) # norm

@memory.cache
def grad_angle(pressure_map):
    x,y = np.gradient(pressure_map)
    return np.arctan2(y,x)
```

```python
feature_generators = {
    'mean':memory.cache(np.nanmean), 
    'max':memory.cache(np.nanmax), 
    'min': memory.cache(np.nanmin),
    'std': memory.cache(np.nanstd),
    'skew': lambda x:scipy.stats.skew(x,axis=None,nan_policy='omit'),
    'kurtosis': lambda x:scipy.stats.kurtosis(x,axis=None,nan_policy='omit'),
    'lower quartile': lambda x:np.nanquantile(x.flatten(),1/4),
    'center quartile': lambda x:np.nanquantile(x.flatten(),2/4),
    'upper quartile': lambda x:np.nanquantile(x.flatten(),3/4),
    'peak count 10': lambda x:count_peaks(x,10),
    'peak count 25': lambda x:count_peaks(x,10),
    'peak count 50': lambda x:count_peaks(x,50),
    'peak count 100': lambda x:count_peaks(x,100),
    'peak count 150': lambda x:count_peaks(x,150),
}

#for i in feature_generators:
#    feature_generators[i] = memory.cache(feature_generators[i])
```

```python
spatial_features = feature_generators.copy()


def gen_features(pressure_map):
    features = []
    # pressure
    for fun in feature_generators.values():
        features.append(fun(pressure_map))
    
    # pressure peaks
    maxima = local_maxima(pressure_map)
    minima = local_minima(pressure_map)
    for fun in feature_generators.values():
        features.append(fun(maxima))
    for fun in feature_generators.values():
        features.append(fun(minima))

    # gradient strenght (~wind)
    magnitude = grad_strength(pressure_map)
    for fun in feature_generators.values():
        features.append(fun(magnitude))
        
    # gradient strenght peaks
    maxima = local_maxima(magnitude)
    minima = local_minima(magnitude)
    for fun in feature_generators.values():
        features.append(fun(maxima))
    for fun in feature_generators.values():
        features.append(fun(minima))

    # gradient direction (~wind)
    angle = grad_angle(pressure_map)
    for fun in feature_generators.values():
        features.append(fun(angle))
        
    # gradient direction peaks
    maxima = local_maxima(angle)
    minima = local_minima(angle)
    for fun in feature_generators.values():
        features.append(fun(maxima))
    for fun in feature_generators.values():
        features.append(fun(minima))

    return features
```

```python
from joblib import delayed,Parallel
```

```python
def calc_features_for_single_ensemble(ensemble_index, ensemble):
    return_value = []
    days = time_mask(np.array(ensemble.time))
    
    for day_index, day in enumerate(days):
        if day_index < 6 or day_index > len(days)-7:
            continue
            
        features = []
        for i in range(-5,6):
            pressure_map = np.array(ensemble.mslp.loc[days[day_index-i],:,:])
            features.extend(gen_features(pressure_map))
                
        return_value.append(features)
    return return_value
```

```python
spatial_data = []

for polex_index, polex in enumerate(simulation_array):
    sim = simulation_array[polex]
    
    tasks = []
    for ensemble_index, ensemble in enumerate((sim)):
        tasks.append(delayed(calc_features_for_single_ensemble)(ensemble_index, ensemble))
        
    spatial_data.append(Parallel(n_jobs=12, verbose=10)(tasks))
        
    
spatial_data = np.array(spatial_data)
```

```python
feature_names = list(feature_generators.keys())
feature_names = ["pressure " + i for i in feature_names] + ["pressure peaks " + i for i in feature_names]+\
["gradient strenght " + i for i in feature_names] + ["gradient strenght peaks " + i for i in feature_names]+\
["gradient direction peaks " + i for i in feature_names]
feature_names = feature_names*11 # ["today's " + i for i in feature_names] + ["yesterday's " + i for i in feature_names] + ["tomorw's " + i for i in feature_names] 
```
